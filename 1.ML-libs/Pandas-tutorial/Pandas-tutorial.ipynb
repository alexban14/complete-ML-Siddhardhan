## Statistical measures
> usecases:
> * finding the mean of x,y,z,.. columns
> * finding the standard deviation of x,y,z,.. columns

***Standard deviation***
> the measurment of how dispersed the data is in relation to the mean

```python
# count / num of values for each column
boston_housing.count()
```

```python
# mean values for each column
boston_housing.mean()
```

```python
# standard deviation for each column
boston_housing.std()
```

```python
# minimum value for each column
boston_housing.min()
```

```python
# maxmimum value for each column
boston_housing.max()
```

```python
# all the statistical measurments
boston_housing.describe()
```

## Manipulating a DataFrame

```python
# adding a column to a DF
californiaHousingDf['Price'] = california_housing.target
```

```python
californiaHousingDf.head()
```

```python
# deleting a row of a DF
californiaHousingDf.drop(index=4, axis=0) # remove row => axis=0
```

```python
# droping a column of a DF
californiaHousingDf.drop(columns='AveBedrms', axis=1) # remove column axis=1
```

```python
# printing particular row using index
californiaHousingDf.iloc[2]
```

```python
# locate particular column
californiaHousingDf.iloc[:, 0] # first column of the DF
```
