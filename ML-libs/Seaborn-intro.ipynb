# Seaborn
> * is a data visualization library based on matplotlib
> * plotting functions that operate on data frames containing whole datasets and internally performing the necessary mapping and statistical aggregation to produce informative plots

```python
import seaborn as sns
import matplotlib as plt
import numpy as np
import pandas as pd
```

## Built-in DataSets in Seaborn

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips') # it will be imported in the form of a Pandas DataFrame
tips.head()
```

```python
# theming the visualization chart
sns.set_theme()
```

```python
# visualizing the data
# loading tips dataset
sns.relplot(
    data = tips,
    x = 'total_bill',
    y = 'tip', col = 'time',
    hue = 'smoker',
    style = 'smoker',
    size = 'size'
)
```

```python
# loading the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

### Scatter Plot

```python
sns.scatterplot(
    x = 'sepal_length',
    y = 'petal_length',
    hue = 'species',
    data = iris
)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

### Count Plot
> * gives us the number of data points for each parameter

```python
sns.countplot(x = 'class', data = titanic)
```

```python
sns.countplot(x = 'survived', data = titanic)
```

### Bar Chart

```python
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic)
```

```python
# california housing price dataset
from sklearn.datasets import fetch_california_housing
california_housing_ds = fetch_california_housing()
californiaHousingDf = pd.DataFrame(california_housing_ds.data, columns = california_housing_ds.feature_names)
californiaHousingDf['Price'] = california_housing_ds.target
californiaHousingDf.head()
```

### Distribution Plot
> * plot containing the range of the values

```python
sns.displot(californiaHousingDf['Price'])
```

### Correlation Matrix:
* Positive Correlation
* Negative Correlation

### Heat Map

```python
correlation = californiaHousingDf.corr()
plt.figure.Figure(figsize = (10, 10))
sns.heatmap(
    correlation,
    cbar = True,
    square = True,
    fmt = '.1f',
    annot = True,
    annot_kws = {'size': 8}, cmap = 'Blues'
)
```
