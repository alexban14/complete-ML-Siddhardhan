# Matplotlib
> * useful for making ***plots*** out of large amounts of data (visual representations of data)
> * each ***pyplot*** function makes a change to a figure (ex: creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # function that takes 100 values evenly spaced between 0-10
y = np.sin(x) # takes x as the angle and it will find the sin value for all the angles
z = np.cos(x) # takes x as the angle and it will find the cosin value for all the angles
```

```python
print(x,y,z)
```

```python
print(y)
```

```python
print(z)
```

## Ploting data

```python
# sin wave
plt.plot(x, y) # plots the graph of x & y
plt.show()
```

```python
# cosin wave
plt.plot(x, z)
plt.show()
```

```python
# adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
# using other symbols for data representation other than a line
plt.plot(x, y, 'r+') # red (color) and + (symbol)
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

## Bar Plot
> * visualizing the magnitude of values

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1]) # incloses the plot in a rectangle
languages = ['python', 'PHP', 'JavaScript', 'Java', 'C#']
programers = [100, 50, 150, 40, 70]
ax.bar(languages, programers)
plt.xlabel('Languages')
plt.ylabel('Number of Programers')
```

## Pie Chart
> * visualizing the distribution of the data in given dataset

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['python', 'PHP', 'JavaScript', 'Java', 'C#']
programers = [100, 50, 150, 40, 70]
ax.pie(programers, labels=languages, autopct='%1.1f%%') # autpct => how many decimal point we need
plt.show()
```

## Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(z)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='g')
plt.show()
```

## 3D Scattered Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x , y, z, c=z, cmap='Blues')
plt.show()
```
