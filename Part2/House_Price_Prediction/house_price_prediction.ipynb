# Boston House Price Prediction Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_df = pd.read_csv('BostonHousing.csv')
print(house_price_df)
```

```python
house_price_df.shape
```

```python
house_price_df.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_df.describe()
```

## Correlation between various features in the dataset

```python
corr = house_price_df.corr()
print(corr)
```

```python
# heatmap to visualize and understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

## Spliting the Features from the Target

```python
X = house_price_df.drop(['price'], axis=1)
Y = house_price_df['price']

print(X, Y)
```

## Train Test data split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

- ***using XGBoost Regressor***

```python
model = XGBRegressor()
```

```python
# training
model.fit(X_train, Y_train)
```

## Model Evaluation

## Predicting on training data

```python
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error: ", score_1)
print("Mean Absolute Error: ", score_2)
```

### Visualizing the actual Prices and the predicted Prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
```

## Predicting on test data

```python
test_data_prediction = model.predict(X_test)

print(test_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error: ", score_1)
print("Mean Absolute Error: ", score_2)
```

### Visualizing the actual Prices and the predicted Prices

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
```
